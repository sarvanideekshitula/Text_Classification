```python
def create_unique_dict(text_arr):
    dictt = {}
    count = -1
    x = np.array(text_arr)
    text1 = (np.unique(x))

    for item in text1:
        data = (item.lower().strip().split(" "))
        for item1 in data:
            if item1 not in dictt:
                count = count +1
                dictt[item1] = count
    print((dictt))
    return dictt
```

```python
def vectorize(text_arr,dictt):
    vectors  = []
    for sentence in text_arr:
        vector = [0] * len(dictt)
        words = (sentence.lower().strip().split(" "))
        for word in words:
            index = dictt[word]
            vector[index] += 1
        vectors.append(vector)
    print(np.array(vectors))
    return vectors
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z)) 
```

```python
def add_intercept(X):
    intercept = np.ones((X.shape[0], 1))
    return np.concatenate((intercept, X), axis=1)
```

```python
def loss(h, y):
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
```

```python
def train(X,y, lr, num_iter):
    X = add_intercept(X)
    theta = np.zeros(X.shape[1])
    print(theta)

    for i in range(num_iter):
        z = np.dot(X, theta)
        h = sigmoid(z)
        gradient = np.dot(X.T, (h - y)) / y.size
        theta -= lr * gradient

        z = np.dot(X, theta)
        h = sigmoid(z)
        los = loss(h, y)
        #print(los)
    return theta
```

```python
def predict(X, theta):
    X = add_intercept(X)
    prob = sigmoid(np.dot(X, theta))
    return prob >= 0.5
```

```python
def evaluate(test_x, test_y, theta):
        y_predicted = predict(test_x, theta)
        correct = 0
        for i,y in enumerate(test_y):
            if y == 0:
                y = False
            else:
                y = True
            if y == y_predicted[i]:
                correct = correct + 1
        total = y_predicted.size

        return (correct/total)*100
```

```python
train(x, y, 0.01, 100)
```

```python
evaluate(x_test, y_test, train(x, y, 0.01, 5000))
```

```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
common_words = get_top_n_words(messages, 20)
```
